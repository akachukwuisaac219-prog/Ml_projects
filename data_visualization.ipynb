# **Data Visualization**

## What is Data Visualization?

Data Visualization refers to the graphical representation of data by visual elements such as charts, infographics, and maps to understand the data. There is a story behind every data, and data visualization brings them to life. It enables decision makers, stakeholders or your readers to see data analytics presented visually, so they can grasp difficult concepts or identify new patterns.  
According to the english adage, a picture is worth a thousand words. Our eyes are easily drawn to colours and patterns. Data visualization helps grab the readers' interest and keeps their on the message of the data.


## Before you visualize your data

1. What relationship am I trying to understand between my data sets?
2. Do I want to understand the distribution of the data or look for outliers?
3. Am I looking to compare multiple values or looking to analyse a single value over time?
4. Am I interested in analyzing trend in my datasets?
5. Is this visualization an important part of my overarching story?

Some important charts include; Bar Charts, pie charts and donout charts for descrete variables. Line graphs, Histograms, scatter plots, ... for continuous variables. Box plots, ...

There are many tools/packages for data visualization in python programming. Some of them are;
1. Matplotlib: helps generate plots, histograms, bar charts, line charts, scatter plots, e.t.c, with just a few line of code.
2. Seaborn: is a python data visualization library based on matplotlib. it produces a high-level interface for drawing sattractive and informative statistical graphics. 
3. Plotly: generate the most interactive graphs; it allows saving them offline and create very rich web-based visualizations.
4. Pandas: also possesses its own data visualization functionalities based on Matplotlib.

## Matplotlib interfaces
Matplotlib has two interfaces.

* **pyplot interface (function-based, implicit):** consists of functions in the pyplot module. Figure and Axes are manipulated through these functions and are only implicitly present in the background.  
API:
**``matplotlib.pyplot``**

````python
plt.plot(x, y)
plt.title("Sample plot")
plt.show()
````

````python
x = np.array([8,6,4,2])
y = np.array([100, 120, 130, 140])
plt.plot(x,y, label= '(x,y)', marker='o', linestyle='--')
plt.plot(2*x,y, label= '(2x,y)', marker='o', linestyle='--')
plt.title('samples')
plt.ylabel('unit')
plt.grid(color='b', linestyle=':')
plt.legend()
plt.show()
````

* **Axes interface (object-based, explicit)**:
  Create a Figure and one or more Axes objects, then explicitly use methods on these objects to add data, configure limits, set labels etc.  
  API:  
**subplots:** create Figure and Axes  
**axes:** add data, limits, labels etc.  
**Figure:** for figure-level methods  
  

````python
# Example - The standard way
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Sample plot")
plt.show()

# OR

# #The old way
fig = plt.figure()
ax = fig.subplots()
ax.plot(x,y)
ax.set_title("Sample plot")
plt.show()
````

````python
fig = plt.figure()
ax = fig.subplots()
x1 = ax.plot(x,y)
x2 = ax.plot(2*x,y)
x3 = ax.plot(x*4,y)
x = ax.plot([1, 2,25,35],[120, 120,120,120])
plt.show()
````

### **Line plots**

**When to Use Line Plots**  
In data analysis, line plots (or line charts) are primarily used to visualize continuity. Use them in:

* **Time Series Analysis:** Tracking a variable over days, months, or years (e.g., stock prices).

* **Trends & Patterns:** Observing if a value is increasing, decreasing, or fluctuating.

* **Relationship between Two Continuous Variables:** Specifically when one variable is dependent on the other (e.g., how "Loss" changes over "Epochs" in Machine Learning).

* **Comparing Series:** Plotting multiple lines to compare different categories (e.g., Salary growth across different Departments).

You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0; therefore, the x data are [0, 1, 2, 3].

plot is a versatile function, and will take an arbitrary number of arguments. For example, to plot x versus y, you can write: 

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
````
