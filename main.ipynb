# **Content based recommendation system**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

## **Feature Selection**

id, title, genre, overview

```python
df = df[["id", "title", "genre", "overview"]]
```

```python
df.head()
```

```python
df["tags"] = df["overview"]+df["genre"]
df.head(1)
```

```python
df.tags
```

```python
df.info()
```

```python
features = df.drop(columns=["genre", "overview"])
features.head()
```

### **Vectorizing the tags**
NB: CountVectorizer has been imported

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
tags_vector = cv.fit_transform(features["tags"].values.astype("U")).toarray()
```

```python
tags_vector
```

```python
import seaborn as sns
```

```python
sns.scatterplot(tags_vector)
```

```python
tags_vector.ndim
```

### **Cosine Similarities**
Nb: ``cosine_similarity`` has been imported

```python
similarity_cos = cosine_similarity(tags_vector)
```

```python
similarity_cos
```

### **Getting films index number**

```python
features[features["title"]=="The Godfather"].index[0]
```

```python
similarity_cos[2]
```

```python
distance = sorted(list(enumerate(similarity_cos[2])), reverse=True, key=lambda tags_vector:tags_vector[1])
distance[0:5]
```

```python
for i in distance[0:5]:
    print(features.iloc[i[0]].title)
```

```python
def recommend(movies):
    try:
        index = features[features["title"]==movies].index[0]
        distance = sorted(list(enumerate(similarity_cos[index])), reverse=True, key=lambda tags_vector:tags_vector[1])
        for i in distance[1:10]:
            print(features.iloc[i[0]].title)
    except:
        print(f'no content on {movies}')
```

```python
recommend("Thor")
```
